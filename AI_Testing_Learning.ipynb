```python
import google.generativeai as genai

# 1. Thiết lập API Key của bạn
from google.colab import userdata
api_key = userdata.get('API_key')
GOOGLE_API_KEY = api_key
genai.configure(api_key=GOOGLE_API_KEY)

# 2. Kiểm tra xem Google đang hỗ trợ những model nào
print("--- DANH SÁCH CÁC MODEL BẠN CÓ THỂ DÙNG ---")
for m in genai.list_models():
    if 'generateContent' in m.supported_generation_methods:
        print(f"Tên model: {m.name}")

# 3. Sau khi thấy danh sách hiện ra, hãy chọn 1 cái tên (thường bắt đầu bằng models/...)
# Ví dụ tôi chọn 'models/gemini-1.5-flash' hoặc 'models/gemini-pro'
SELECTED_MODEL = 'models/gemini-1.5-flash' # Bạn hãy thay tên khớp với danh sách in ra ở trên

model = genai.GenerativeModel(SELECTED_MODEL)

# 4. Chạy thử nghiệm lại
questions = ["Thủ đô Việt Nam là gì?"]
for q in questions:
    try:
        response = model.generate_content(q)
        print(f"\nCâu hỏi: {q}")
        print(f"Kết quả thành công: {response.text}")
    except Exception as e:
        print(f"\nVẫn còn lỗi: {e}")
```

```python
import google.generativeai as genai

# Thiết lập API Key (Nhớ dán lại key của bạn)
from google.colab import userdata
api_key = userdata.get('API_key')
GOOGLE_API_KEY = api_key
genai.configure(api_key=GOOGLE_API_KEY)

# SỬA Ở ĐÂY: Thêm chữ '-latest' hoặc 'models/' phía trước
model = genai.GenerativeModel('gemini-2.5-flash')

questions = [
    "Thủ đô của Việt Nam là gì?"
]

print("--- BẮT ĐẦU CHẠY PIPELINE TEST ---")

for q in questions:
    try:
        response = model.generate_content(q)
        print(f"Câu hỏi: {q}")
        print(f"AI trả lời: {response.text}")
        print("-" * 20)
    except Exception as e:
        print(f"Lỗi tại câu hỏi '{q}': {e}")

print("--- HOÀN THÀNH ---")
```
