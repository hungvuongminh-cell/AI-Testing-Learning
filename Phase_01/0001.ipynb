```python
from google import genai
from google.colab import userdata
import time

# 1. API CONFIGURATION (New SDK Style)
api_key = userdata.get('key')
client = genai.Client(api_key=api_key)

# 2. MODEL SELECTION
# Note: Using the latest stable model
SELECTED_MODEL = 'gemini-2.5-flash'

# 3. TEST DATASET
questions = [
    "What is the capital of Vietnam?",
    "What is 1 + 1?",
    "Name a famous dish in Ho Chi Minh City"
]

print("--- STARTING TEST EXECUTION (NEW SDK) ---")

# 4. EXECUTION LOOP
for q in questions:
    try:
        # New method: models.generate_content
        response = client.models.generate_content(
            model=SELECTED_MODEL,
            contents=q
        )

        print(f"Question: {q}")
        print(f"AI Response: {response.text.strip()}")
        print("-" * 20)

        # Adding a small delay to respect free tier quotas
        time.sleep(10)

    except Exception as e:
        print(f"Error encountered: {e}")

print("--- TEST EXECUTION COMPLETED ---")
```
