```python
import google.generativeai as genai

# 1. THIẾT LẬP API KEY
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('key')
genai.configure(api_key=GOOGLE_API_KEY)

# 2. CHỌN LUÔN MODEL (Bỏ qua bước quét danh sách đang bị lỗi Unicode)
# 'gemini-2.5-flash' là model tiêu chuẩn nhất hiện nay
SELECTED_MODEL = 'models/gemini-2.5-flash'
model = genai.GenerativeModel(SELECTED_MODEL)

# 3. BỘ ĐỀ THI
questions = [
    "Thủ đô Việt Nam là gì?",
    "1 + 1 bằng mấy?",
    "Kể tên một món ăn ngon ở Sài Gòn"
]

print("--- ĐANG CHẠY TEST ---")

for q in questions:
    try:
        response = model.generate_content(q)
        print(f"Hỏi: {q}")
        print(f"AI đáp: {response.text}")
        print("-" * 20)
    except Exception as e:
        print(f"Lỗi: {e}")

print("--- HOÀN THÀNH ---")
```
