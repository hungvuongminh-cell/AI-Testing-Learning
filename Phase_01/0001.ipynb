```python
import google.generativeai as genai
from google.colab import userdata

# 1. API CONFIGURATION
# Fetching the API Key securely from Colab Secrets
GOOGLE_API_KEY = userdata.get('key')
genai.configure(api_key=GOOGLE_API_KEY)

# 2. MODEL SELECTION
# Note: If the selected model is unavailable, switch to an alternative (e.g., 'gemini-1.5-flash')
SELECTED_MODEL = 'models/gemini-2.5-flash'
model = genai.GenerativeModel(SELECTED_MODEL)

# 3. TEST DATASET
# Defining the list of prompts to evaluate the AI
questions = [
    "What is the capital of Vietnam?",
    "What is 1 + 1?",
    "Name a famous dish in Ho Chi Minh City"
]

print("--- STARTING TEST EXECUTION ---")

# 4. EXECUTION LOOP
for q in questions:
    try:
        # Sending request to the AI model
        response = model.generate_content(q)

        print(f"Question: {q}")
        print(f"AI Response: {response.text.strip()}")
        print("-" * 20)

    except Exception as e:
        # Error handling to prevent the pipeline from crashing
        print(f"Error encountered: {e}")

print("--- TEST EXECUTION COMPLETED ---")
```
