```python
import google.generativeai as genai
import time # Library to control time delays

# 1. API KEY SETUP
from google.colab import userdata
userdata.get('secretName')
GOOGLE_API_KEY = 'secretName'
genai.configure(api_key=GOOGLE_API_KEY)

# 2. DATASET (TEST CASE)
# Reduced to only one question as requested
questions = [
    "What is the capital of Vietnam?"
]

print("--- STARTING TEST ON ALL AVAILABLE MODELS ---")

# 3. OUTER LOOP: Automatically retrieve model list from the system
for m in genai.list_models():
    # Check if the model supports text generation (generateContent)
    if 'generateContent' in m.supported_generation_methods:

        current_model_name = m.name # Get model name (e.g., models/gemini-1.5-flash)
        print(f"\n--- TESTING MODEL: {current_model_name} ---")

        try:
            # Initialize the selected model
            model = genai.GenerativeModel(current_model_name)

            # 4. INNER LOOP: Run the question through the current model
            for q in questions:
                response = model.generate_content(q)
                print(f"Question: {q}")
                print(f"AI Response: {response.text.strip()}") # .strip() removes extra whitespaces
                print("-" * 10)

            # ADDED: 10-second break after finishing each model to avoid rate limits
            print("Finished model test. Waiting 10 seconds before the next one...")
            time.sleep(10)

        except Exception as e:
            # Handle errors for legacy or specialized models to keep the pipeline running
            print(f"Model {current_model_name} encountered an error: {e}")

print("\n--- ALL MODELS HAVE COMPLETED THE TEST ---")
```
